```python
import yfinance as yf
import datetime
from yahoo_fin.stock_info import get_data
import matplotlib.pyplot as plt
import pandas as pd
from pandas.plotting import scatter_matrix
import mpl_finance
from mpl_finance import candlestick_ohlc
from matplotlib.dates import DateFormatter, date2num, WeekdayLocator, DayLocator, MONDAY
```

```python
start = datetime.datetime(2013,1,1)
end = datetime.datetime(2023,12,31)
```

```python
Google = yf.download("GOOGL", start=start, end=end)
Apple = yf.download("AAPL", start=start, end=end)
Microsoft = yf.download("MSFT", start=start, end=end)
Nvidia = yf.download("NVDA", start=start, end=end)
Amazon =yf.download("AMZN", start=start, end=end)
```

```python
# Visual representation of total amount of money being traded (market capital)
Google["Total Traded"] = Google["Open"] * Google["Volume"]
Apple["Total Traded"] = Apple["Open"] * Apple["Volume"]
Microsoft["Total Traded"] = Microsoft["Open"] * Microsoft["Volume"]
Nvidia["Total Traded"] = Nvidia["Open"] * Nvidia["Volume"]
Amazon["Total Traded"] = Amazon["Open"] * Amazon["Volume"]
```

```python
Google["Total Traded"].plot(label= "Google", figsize=(15,7))
Apple["Total Traded"].plot(label= "Apple", figsize=(15,7))
Microsoft["Total Traded"].plot(label= "Microsoft", figsize=(15,7))
Nvidia["Total Traded"].plot(label= "Nvidia", figsize=(15,7))
Amazon["Total Traded"].plot(label= "Amazon", figsize=(15,7))
plt.legend()
plt.title("Total Traded")

##Huge amount of money traded for Nvidia around 2023
```
