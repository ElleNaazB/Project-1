```python
# CALCULATION OF EQUITY FOR TOP 5 COMPANIES

# List of ticker symbols for the companies
ticker_symbols = ["AAPL", "GOOGL", "AMZN", "MSFT", "NVDA"]

# Define the equity components to sum up for Total Equity
equity_components = ['Common Stock Equity']

# Define the years to loop through
years = ['2023', '2022', '2021', '2020']

for symbol in ticker_symbols:
    print(f"Calculating Total Equity for {symbol}:")
    
    # Initialize a dictionary to store total equity for each year
    total_equity_by_year = {year: 0 for year in years}
    
    # Fetch balance sheet data for the company
    ticker = yf.Ticker(symbol)
    balance_sheet = ticker.balance_sheet
    
    # Loop through each equity component and sum up the values for each year
    for year in years:
        if year in balance_sheet.columns:
            for equity_component in equity_components:
                total_equity_by_year[year] += balance_sheet.loc[equity_component, year]
    
    # Print the total equity for each year
    for year in years:
        if year in balance_sheet.columns and not balance_sheet.loc[equity_components[0], year].empty:
            total_equity = total_equity_by_year[year].values[0]
        else:
            total_equity = "No data found for the year"
        print(f"Total Equity for {year} for {symbol}: {total_equity}")
```

```python
# CALCULATION OF DEBT FOR THE TOP 5 COMPANIES
# List of ticker symbols for the companies
ticker_symbols = ["AAPL", "GOOGL", "AMZN", "MSFT", "NVDA"]

# Define the debt components to sum up for Total Debt
debt_components = ['Total Debt']

# Define the years to loop through
years = ['2023', '2022', '2021', '2020']

for symbol in ticker_symbols:
    print(f"Calculating Total Debt for {symbol}:")
    
    # Initialize a dictionary to store total debt for each year
    total_debt_by_year = {year: 0 for year in years}
    
    # Fetch balance sheet data for the company
    ticker = yf.Ticker(symbol)
    balance_sheet = ticker.balance_sheet
    
    # Loop through each debt component and sum up the values for each year
    for year in years:
        if year in balance_sheet.columns:
            for debt_component in debt_components:
                total_debt_by_year[year] += balance_sheet.loc[debt_component, year]
    
    # Print the total debt for each year
    for year in years:
        total_debt = total_debt_by_year[year].values[0] if year in balance_sheet.columns and not balance_sheet.loc[debt_components[0], year].empty else "No data found for the year"
        print(f"Total Debt for {year} for {symbol}: {total_debt}")
```

```python
# CALCULATION OF DEBT TO EQUITY RATIO

# List of ticker symbols for the companies
ticker_symbols = ["AAPL", "GOOGL", "AMZN", "MSFT", "NVDA"]

# Define the equity components to sum up for Total Equity
equity_components = ['Common Stock Equity']

# Define the debt components to sum up for Total Debt
debt_components = ['Total Debt']

# Define the years to loop through
years = ['2023', '2022', '2021', '2020']

for symbol in ticker_symbols:
    print(f"Calculating Debt to Equity Ratio for {symbol}:")
    
    # Initialize dictionaries to store total equity and total debt for each year
    total_equity_by_year = {year: 0 for year in years}
    total_debt_by_year = {year: 0 for year in years}
    
    # Fetch balance sheet data for the company
    ticker = yf.Ticker(symbol)
    balance_sheet = ticker.balance_sheet
    
    # Loop through each equity component and sum up the values for each year
    for year in years:
        if year in balance_sheet.columns:
            for equity_component in equity_components:
                total_equity_by_year[year] += balance_sheet.loc[equity_component, year]
    
    # Loop through each debt component and sum up the values for each year
    for year in years:
        if year in balance_sheet.columns:
            for debt_component in debt_components:
                total_debt_by_year[year] += balance_sheet.loc[debt_component, year]
    
    # Calculate and print the Debt to Equity Ratio for each year
    for year in years:
        total_debt_series = total_debt_by_year[year]
        total_equity_series = total_equity_by_year[year]
        
        if not total_debt_series.empty and not total_equity_series.empty:
            total_debt = total_debt_series.values[0]
            total_equity = total_equity_series.values[0]
            
            if total_equity != 0:
                debt_to_equity_ratio = total_debt / total_equity
                print(f"Debt to Equity Ratio for {year} for {symbol}: {debt_to_equity_ratio:.2f}")
            else:
                print(f"Cannot calculate Debt to Equity Ratio for {year} for {symbol}: Total Equity is zero")
        else:
            print(f"No data available for Debt to Equity Ratio calculation for {year} for {symbol}")

```

```python
# COLLATE CALCULATED EQUITY, DEBT and DEBT TO EQUITY RATIO INTO A TABLE 
 
# List of ticker symbols for the companies
ticker_symbols = ["AAPL", "GOOGL", "AMZN", "MSFT", "NVDA"]

# Define the equity components to sum up for Total Equity
equity_components = ['Common Stock Equity']

# Define the debt components to sum up for Total Debt
debt_components = ['Total Debt']

# Define the years to loop through
years = ['2023', '2022', '2021', '2020']

# Initialize a dictionary to store the calculated values
debt_equity_data = {
    'Company': [],
    'Year': [],
    'Total Equity': [],
    'Total Debt': [],
    'Debt to Equity Ratio': []
}

for symbol in ticker_symbols:
    # Fetch balance sheet data for the company
    ticker = yf.Ticker(symbol)
    balance_sheet = ticker.balance_sheet
    
    for year in years:
        total_equity = 0
        total_debt = 0
        debt_to_equity_ratio = 'N/A'
        
        if year in balance_sheet.columns:
            for equity_component in equity_components:
                equity_values = balance_sheet.loc[equity_component, year]
                if not equity_values.empty:
                    total_equity += equity_values.values[0]
            
            for debt_component in debt_components:
                debt_values = balance_sheet.loc[debt_component, year]
                if not debt_values.empty:
                    total_debt += debt_values.values[0]
            
            if total_equity != 0:
                debt_to_equity_ratio = total_debt / total_equity
        
        debt_equity_data['Company'].append(symbol)
        debt_equity_data['Year'].append(year)
        debt_equity_data['Total Equity'].append(total_equity)
        debt_equity_data['Total Debt'].append(total_debt)
        debt_equity_data['Debt to Equity Ratio'].append(debt_to_equity_ratio)

# Create a DataFrame from the dictionary
debt_equity_data_df = pd.DataFrame(debt_equity_data)

# Set display options to show complete numerical values
pd.set_option('display.float_format', '{:.2f}'.format)

# Display the table with complete numerical values
debt_equity_data_df
```

```python
# REFORMAT THE TABLE BY COMPANY COLUMNS

# List of ticker symbols for the companies
ticker_symbols = ["AAPL", "GOOGL", "AMZN", "MSFT", "NVDA"]

# Define the equity components to sum up for Total Equity
equity_components = ['Common Stock Equity']

# Define the debt components to sum up for Total Debt
debt_components = ['Total Debt']

# Define the years to loop through
years = ['2023', '2022', '2021', '2020']

# Initialize a dictionary to store the calculated values
debt_equity_data = {
    'Company': [],
    'Year': [],
    'Metric': [],
    'Value': []
}

for symbol in ticker_symbols:
    # Fetch balance sheet data for the company
    ticker = yf.Ticker(symbol)
    balance_sheet = ticker.balance_sheet
    
    for year in years:
        total_equity = 0
        total_debt = 0
        debt_to_equity_ratio = 'N/A'
        
        if year in balance_sheet.columns:
            for equity_component in equity_components:
                equity_values = balance_sheet.loc[equity_component, year]
                if not equity_values.empty:
                    total_equity += equity_values.values[0]
            
            for debt_component in debt_components:
                debt_values = balance_sheet.loc[debt_component, year]
                if not debt_values.empty:
                    total_debt += debt_values.values[0]
            
            if total_equity != 0:
                debt_to_equity_ratio = total_debt / total_equity
        
        debt_equity_data['Company'].append(symbol)
        debt_equity_data['Year'].append(year)
        debt_equity_data['Metric'].append('Total Equity')
        debt_equity_data['Value'].append(total_equity)
        
        debt_equity_data['Company'].append(symbol)
        debt_equity_data['Year'].append(year)
        debt_equity_data['Metric'].append('Total Debt')
        debt_equity_data['Value'].append(total_debt)
        
        debt_equity_data['Company'].append(symbol)
        debt_equity_data['Year'].append(year)
        debt_equity_data['Metric'].append('Debt to Equity Ratio')
        debt_equity_data['Value'].append(debt_to_equity_ratio)

# Create a DataFrame from the dictionary
debt_equity_data_df = pd.DataFrame(debt_equity_data)

# Pivot the DataFrame to reformat the table
pivot_df = debt_equity_data_df.pivot(index=['Year', 'Metric'], columns='Company', values='Value').reset_index()

# Set display options to show complete numerical values
pd.set_option('display.float_format', '{:.2f}'.format)

# Display the restructured table
pivot_df
```
