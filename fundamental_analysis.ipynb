```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
yf.Ticker("AMZN").info
```

```python
tickers = ["AAPL", "MSFT", "NVDA", "AMZN", "GOOGL"]
```

```python
infos = []
for ticker in tickers: 
    infos.append(yf.Ticker(ticker).info)
```

```python
fundamentals = ['dividendYield', 'marketCap', 'beta', 'forwardPE']
```

```python
df = pd.DataFrame(infos).set_index("symbol")
df
```

```python
#filter the data set 
filtered_data = df[df.columns[df.columns.isin(fundamentals)]]
filtered_data
```

```python
plt.bar(df.index, df.dividendYield, color = ["red", "blue", "green","orange", "yellow"])
df.dividendYield.nlargest(5)
```
