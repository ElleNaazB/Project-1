```python
import yfinance as yf
import matplotlib.pyplot as plt
from yahoo_fin.stock_info import get_data

# Download historical data for Apple stock
aapl = yf.download('AAPL', start='2010-01-01', end='2023-01-01')

# Plot the closing price
aapl['Close'].plot(figsize=(14,7), title='Apple Stock Price')

# Calculate and plot moving averages for different durations
aapl['SMA_50'] = aapl['Close'].rolling(window=50).mean()  # Short-term trend
aapl['SMA_200'] = aapl['Close'].rolling(window=200).mean()  # Long-term trend

aapl['SMA_50'].plot(label='50-Day SMA')
aapl['SMA_200'].plot(label='200-Day SMA')

```

```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Download historical stock prices
data = yf.download('AAPL', start='2010-01-01', end='2023-01-01')

# Prepare the features (X) and target (y)
# We'll use the number of days since the start of the series as our feature
X = np.array(range(len(data))).reshape(-1, 1)
y = data['Close'].values

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Make predictions
predictions = model.predict(X)

# Plot the historical data
data['Close'].plot(figsize=(14,7), label='Historical Daily Closing Price')

# Plot the regression line
plt.plot(data.index, predictions, color='red', label='Regression Line')

# Show the plot
plt.legend()
plt.title('Linear Regression on AAPL Stock Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```
