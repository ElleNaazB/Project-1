
# Which of the top 5 S&P 500 companies had the highest return on investment over the selected time period?

 what is Cumulative return? 
 Cumulative return is the sums up of the total return over a specified period.
 fot calculiting of cumulative return will be used this formula:

 Daily percentage change: 
 r_t=p_t/p_(t-1) -1

 Cumulative Retuern:
 i_t=(1+r_t ) i_(t-1)=(1+p_t/p_(t-1) -1) i_(t-1)=p_t/p_(t-1)  i_(t-1)

```python
# finding the Cunulative return/totla investment return 
Apple_weekly['Cumulative Return'] = (1 + Apple_weekly['returns']).cumprod()
Apple_weekly.head()
```

```python
NVIDIA_weekly['Cumulative Return'] = (1 + NVIDIA_weekly['returns']).cumprod()
NVIDIA_weekly.head()
```

```python
GOOGL_weekly['Cumulative Return'] = (1 + GOOGL_weekly['returns']).cumprod()
GOOGL_weekly.head()
```

```python
amazon_weekly['Cumulative Return'] = (1 + amazon_weekly['returns']).cumprod()
amazon_weekly.head()
```

```python
msft_weekly['Cumulative Return'] = (1 + msft_weekly['returns']).cumprod()
msft_weekly.head()
```

```python
# ploting of the highest investment returns companies 
Apple_weekly['Cumulative Return'].plot( label= 'AAPL', figsize=(15,7))
amazon_weekly['Cumulative Return'].plot(label= 'AMZN')
NVIDIA_weekly['Cumulative Return'].plot(label= 'NVDA')
msft_weekly['Cumulative Return'].plot(label= 'MSFT')
GOOGL_weekly['Cumulative Return'].plot(label= 'GOOGL')
plt.title('Cumulative Returns for Top 5 S&P 500 Companies')
plt.xlabel('Date')
plt.ylabel('Cumulative Return')
plt.legend()
plt.grid(True)
plt.show()
```
