How have the stock prices of the top 5 S&P 500 companies trended over the past years?

Nvidia's stock, represented by the blue line, stands out as the leader in growth throughout the period analyzed. It indicates that Nvidia's stock significantly outpaced others in terms of overall returns, particularly surging after 2021, followed by a notable decline in 2022 and subsequent recovery.

Nvidia, a major chip manufacturer in the United States, experienced a steep drop in its stock price post-2022 due to newly imposed export restrictions by the US government on the company's artificial intelligence (AI) chips destined for China and Russia. This led to Nvidia hitting a 52-week low in its stock value.

On the other hand, Microsoft's stock, depicted by the green line, shows a more gradual incline over the decade with fewer dramatic spikes compared to Nvidia.

Nvidia, like any tech company, has seen fluctuations in the stock market, but its success story through 2023 stands out among its peers. As the leading producer of GPUs worldwide, Nvidia consistently surpassed earnings expectations throughout the year, earning it recognition as one of the top-performing S&P stocks by the end of 2023.

The primary driver behind Nvidia's success lies in the surging interest in AI, generative AI, and LLMs (large language models). From 2006 to 2023, Apple's revenue has seen significant fluctuations and growth, reflecting the company's performance and market dynamics. Apple’s revenue has been over 100 billion ever since 2011 and almost doubled by 2014, tripled by 2021. Notably, there have been periods of rapid expansion, such as the years following the introduction of groundbreaking products like the iPhone and iPad, while other years have seen more modest growth or slight declines. Overall, Apple's revenue trajectory showcases its evolution as a major player in the technology industry.


Since its launch in 2008, the iPhone has remained Apple's flagship product and primary revenue generator. Even with the introduction of innovative products like the Apple Watch and AirPods, the iPhone has maintained its dominance, still contributing to half of the company's revenue.

Despite its enduring global popularity, the iPhone's revenue trajectory has not been consistently upward. Apple experienced double-digit year-on-year revenue growth for the iPhone from 2008 to 2015. However, revenue growth stagnated from 2015 to 2021.

In the most recent year, there was a notable breakthrough for the iPhone, with a significant surge in sales leading to a 39% increase in revenue. This performance suggests a potential positive impact on Apple's stock price, as strong iPhone sales typically correlate with favorable financial results and investor confidence. However, other factors, such as overall market conditions and the performance of other Apple products and services, also influence the stoc and total volume tradedk price.

```python
#Calculation for total  money traded by each company

Apple['Total Traded'] = Apple['Open']*Apple['Volume']
Google['Total Traded'] =Google['Open']*Google['Volume']
Nvidia['Total Traded'] = Nvidia['Open']*Nvidia['Volume']
Microsoft['Total Traded'] = Microsoft['Open']*Microsoft['Volume']
Amazon['Total Traded'] = Amazon['Open']*Amazon['Volume']
```
