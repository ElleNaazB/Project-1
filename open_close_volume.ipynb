```python
import yfinance as yf
import datetime
from yahoo_fin.stock_info import get_data
import matplotlib.pyplot as plt
import pandas as pd
```

```python
start = datetime.datetime(2013,1,1)
end = datetime.datetime(2023,12,31)
```

```python
tickers = ["GOOGL", "AAPL", "MSFT", "NVDA", "AMZN" ]
company_info = []
for ticker in tickers:
    df = yf.download(ticker, start=start, end=end)
    df['Symbol'] = ticker  # Add a column with the ticker symbol
    company_info.append(df)
```

```python
info = pd.concat(company_info, axis=0)
info
```

```python


# Set the figure size for better visibility
plt.figure(figsize=(17, 9))

# Define the number of subplots, which is 5 in this case (for Open, High, Low, Close, and Adj Close)
num_plots = 5

# Create a figure and a grid of subplots
fig, axs = plt.subplots(num_plots, 1, figsize=(15, 25))

# Titles for each subplot
titles = ['Open Price', 'High Price', 'Low Price', 'Close Price', 'Adj Close Price']

# Plot data for each company in a separate subplot
for i, company in enumerate(company_info):
    # Get the company symbol from the DataFrame
    symbol = company['Symbol'].iloc[0]
    
    # Plot each price type in its subplot
    axs[0].plot(company['Open'], label=f'{symbol} Open')
    axs[1].plot(company['High'], label=f'{symbol} High')
    axs[2].plot(company['Low'], label=f'{symbol} Low')
    axs[3].plot(company['Close'], label=f'{symbol} Close')
    axs[4].plot(company['Adj Close'], label=f'{symbol} Adj Close')

# Set titles, legends, and labels for each subplot
for ax, title in zip(axs, titles):
    ax.set_title(title)
    ax.legend()
    ax.set_ylabel('Price')
    ax.set_xlabel('Date')

plt.tight_layout()  # Adjust the layout to prevent overlap
plt.show()  # Display the plots
```

```python

# Assuming company_info is a list of DataFrames and each DataFrame has a 'Symbol' column
for df in company_info:
    symbol = df['Symbol'].iloc[0]  # Get the company symbol from the 'Symbol' column
    plt.figure(figsize=(17,9))
    
    # Plot each price type
    plt.plot(df['Open'], label='Open Price')
    plt.plot(df['High'], label='High Price')
    plt.plot(df['Low'], label='Low Price')
    plt.plot(df['Close'], label='Close Price')
    plt.plot(df['Adj Close'], label='Adj Close Price')
    
    # Add titles and labels
    plt.title(f'{symbol} Stock Prices')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    
    # Show the plot
    plt.show()
```
