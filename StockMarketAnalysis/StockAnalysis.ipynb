```python
import datetime
import yfinance as yf
import pandas as pd
import matplotlib as plt
import numpy as np
import matplotlib.pyplot as plt
```

```python
start = datetime.datetime(2013,1,1)
end = datetime.datetime(2023,12,31)
```

```python
Google = yf.download("GOOGL", start=start, end=end)
Apple = yf.download("AAPL", start=start, end=end)
Microsoft = yf.download("MSFT", start=start, end=end)
Nvidia = yf.download("NVDA", start=start, end=end)
Amazon =yf.download("AMZN", start=start, end=end)
```

```python
Nvidia['Open'].plot(label='Nvda',figsize=(17,7))
Amazon['Open'].plot(label='Amazon')
Microsoft['Open'].plot(label='Msft')
Apple['Open'].plot(label='Apple')
Google['Open'].plot(label='Google')
plt.ylabel('Stock Price')
plt.title('Stock Prices of Top 5 companies')
plt.legend()
```

```python
Nvidia['Volume'].plot(label='Nvda', figsize=(15,7))
Microsoft['Volume'].plot(label='Msft', figsize=(15,7))
Google['Volume'].plot(label='Google')
Amazon['Volume'].plot(label='Amazon')
Apple['Volume'].plot(label='Apple')
plt.ylabel('Volume Traded')
plt.legend()
plt.show()
```

```python
Apple['Open'].plot(label='Apple Open Price',figsize=(17,9))
Apple['Close'].plot(label='Apple Close Price')
Apple['High'].plot(label='Apple High Price')
Apple['Low'].plot(label='Apple Low Price')
plt.legend()
plt.title('Apple Stock Price')
plt.ylabel('Stock Price')
```

```python
Nvidia['Volume'].plot(label='Nvda', figsize=(15,7))
Apple['Volume'].plot(label='Apple')
Microsoft['Volume'].plot(label='Msft')
Google['Volume'].plot(label='Google')
Amazon['Volume'].plot(label='Amazon')
plt.ylabel('Volume Traded')
plt.legend()
plt.show()
```
